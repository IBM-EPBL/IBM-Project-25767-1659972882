```python
quartile1=data['Rings'].quantile(0.25)
quartile2=data['Rings'].quantile(0.75)
Interquartile_range=quartile2-quartile1
whisker_width = 1.5
lower_whisker = quartile1 -(whisker_width*Interquartile_range)
upper_whisker = quartile2 + (whisker_width*Interquartile_range)
data['Rings']=np.where((data['Rings'])>upper_whisker,upper_whisker,np.where(data['Rings']<lower_whisker,lower_whisker,data['Rings']))
```

<h2> 10. Split the data into training and testing</h2>

```python
to_numeric = {'M': 1.0, 'F': 2.0,
'I': 0.0}
df = data.applymap(lambda lable: to_numeric.get(lable) if lable in to_numeric else lable)
Gender = pd.to_numeric(df.Sex)
df.drop(['Sex'], axis = 1, inplace = True)
df = pd.concat([df, Gender], axis = 1)
print(f"training set (row, col): {df.shape}\n")
```

```python
df.info()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

y = df['Rings']
X = df.drop('Rings', axis = 1)
 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

<h2>11. Build the Model</h2>

```python
linear_regression = LinearRegression()
```

<h2> 12. Train the Model</h2>

```python
linear_regression.fit(X_train, y_train)
```

<h2> 13. Test the Model</h2>

```python
y_predict = linear_regression.predict(X_test)
```

<h2> 14. Measure the performance using Metrics.</h2>

```python
linear_regression.score(X_test, y_test)
```
